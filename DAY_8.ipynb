## Analyze seasonal results by method

### Subtask:
Calculate and display the number of matches won by runs and by wickets for each season.


**Reasoning**:
Group the data by season and result, count the occurrences, unstack the table, and fill missing values with 0 to show wins by runs and wickets per season.



```python
season_result_wins = match_data.groupby(['season', 'result']).size().unstack(fill_value=0)
display(season_result_wins)
```

## Analyze virat kohli's player of the match awards

### Subtask:
Compare the number of Player of the Match awards won by Virat Kohli when batting first versus batting second.


**Reasoning**:
Filter the dataframe for Virat Kohli's Player of the Match awards and then count wins by runs and wickets.



```python
kohli_potm = match_data[match_data['player_of_match'] == 'V Kohli']
kohli_batting_first_wins = kohli_potm[kohli_potm['result'] == 'runs'].shape[0]
kohli_batting_second_wins = kohli_potm[kohli_potm['result'] == 'wickets'].shape[0]

print(f"Virat Kohli Player of the Match awards won batting first (by runs): {kohli_batting_first_wins}")
print(f"Virat Kohli Player of the Match awards won batting second (by wickets): {kohli_batting_second_wins}")
```

```python
top_venues=match_data["venue"].value_counts().head(5).index
print(top_venues)
venue_team_wins=match_data.groupby(['venue',"winner"]).size().unstack(fill_value=0).loc[top_venues]
venue_team_wins.T
```

```python
#plot as a stacked bar graph
venue_team_wins.T.plot(kind='bar',stacked=True,figsize=(14,6),colormap="plasma")
plt.title('match won by teams at Top 5 Venues')
plt.xlabel('teams')
plt.ylabel('wins')
plt.legend(title='Venue',loc='upper left')
plt.tight_layout()
plt.show()
```

```python
#reset_index converts the series index to dataframe
#rename is used to change column name
seasongand=match_data.groupby(['season','toss_decision']).size().reset_index().rename(columns={0:'count'})
seasongand
```

```python
plt.figure(figsize=(10,6))
ax = sns.barplot(x="season",y='count',hue='toss_decision',data=seasongand)
#plotting barplot of 'Season'vs"count"for bat & field both....
plt.title('Toss Decision by Season')
plt.xlabel('Season')
plt.ylabel('Count')
plt.legend(title='Toss Decision')

for container in ax.containers:
    ax.bar_label(container)

plt.tight_layout()
plt.show()
```
