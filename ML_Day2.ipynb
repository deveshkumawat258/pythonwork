```python
import pandas as pd
import numpy as np
df=pd.read_csv("/content/covid_toy - covid_toy.csv")
df.head(3)
```

```python
df.isnull().sum()  # for chercking missing values
```

```python
from sklearn.impute import SimpleImputer ##for fill missing values
```

```python
si=SimpleImputer(strategy='mean')
```

```python
df["fever"]=si.fit_transform(df[['fever']])  #fit means learn the parameters anad transform means to apply it
```

```python
#df.isnull().sum()
```

```python
df.head(2)
#df['has_covid']
```

```python
df['gender']=df['gender'].map({'Female':0,"Male":1})
df['cough']=df['cough'].map({'Mild':0,"Strong":1})
df['city']=df['city'].map({'Kolkata':0,"Bangalore":1,"Delhi":2,"Mumbai":3})
df['has_covid']=df['has_covid'].map({'No':0,"Yes":1})
```

```python
df.head(3)
```

```python
x=df.drop(columns=["has_covid"])#input data
y=df["has_covid"]#target data
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print("total data shape:",df.shape)
print("------------------------------")
print("Input data shape:",x.shape)
print("x_train data shape",x_train.shape)
print("x_test data shape",x_test.shape)
print("------------------------------")
print("Target data shape:",y.shape)
print("y_train data shape;",y_train.shape)
print("y_test data shape;",y_test.shape)
```

```python
np.round(x_train.describe(),2)
```

```python
#standardisation
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train_sc=sc.fit_transform(x_train)  #fit is to learn the parameter and transform to apply it
```

```python
x_train_new=pd.DataFrame(x_train_sc,columns=x_train.columns)
np.round(x_train_new.describe(),2)
```
